```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator
import matplotlib.dates as mdates
import seaborn as sns
```

```python
hx = pd.read_csv('Excel_Files/Complete_Data/HX_Data.csv')
hx.head()
```

```python
#Calculate Sink Rates
def Divide(a,b):
    if b == 0:
        return np.nan
    return a / b
hx['Overall Sink Rate'] = hx.apply(lambda row: Divide(row['Number of Ships Sunk'], row['Number of Ships'], ), axis=1)
hx['Escort Sink Rate'] = hx.apply(lambda row: Divide(row['Number of Escorts Sunk'], row['Number of Escort Ships']), axis=1)
hx['Straggler Sink Rate'] = hx.apply(lambda row: Divide(row['Number of Stragglers Sunk'], row['Number of Stragglers'],), axis=1)
hx['Depart_Date'] = pd.to_datetime(hx['Depart_Date'], errors='coerce')
hx = hx.fillna(0)
hx.head()
```
